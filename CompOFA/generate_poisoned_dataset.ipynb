```python
import os
import shutil
import random
```

```python
split_percentage = 0.3

# Clean up old poisoned data set to create new one with new split
poisoned_data_path = '/home/cloud/Desktop/abhi/VillainNet/CompOFA/poisoned_data/'

if os.path.exists(poisoned_data_path):
    shutil.rmtree(poisoned_data_path)

os.makedirs(poisoned_data_path)

sub_directories = ['train', 'val']
labels = ['background', 'pedestrianCrossing', 'speedLimit', 'stop']
for sub in sub_directories:
    sub_path = os.path.join(poisoned_data_path, sub)
    os.makedirs(sub_path)
    for label in labels:
        full_path = os.path.join(sub_path, label)
        os.makedirs(full_path)

base_data_path = '/home/cloud/Desktop/abhi/VillainNet/CompOFA/data/'
just_poisoned_data_path = '/home/cloud/Desktop/abhi/VillainNet/CompOFA/just_poisoned_data/'

for sub in sub_directories:
    sub_src_path = os.path.join(base_data_path, sub)
    sub_dst_path = os.path.join(poisoned_data_path, sub)
    sub_just_poisoned_path = os.path.join(just_poisoned_data_path, sub)
    for label in labels:
        full_src_path = os.path.join(sub_src_path, label)
        full_dst_path = os.path.join(sub_dst_path, label)
        full_just_poisoned_path = os.path.join(sub_just_poisoned_path, label)

        base_data_files = [os.path.join(full_src_path, f) for f in os.listdir(full_src_path) if os.path.isfile(os.path.join(full_src_path, f))]
        just_poisoned_files = [os.path.join(full_just_poisoned_path, f) for f in os.listdir(full_just_poisoned_path) if os.path.isfile(os.path.join(full_just_poisoned_path, f))]
        
        random.shuffle(base_data_files)
        random.shuffle(just_poisoned_files)
        
        num_base_files = len(base_data_files)
        num_poisoned_files = len(just_poisoned_files)

        # # Calculate the maximum possible size of the final dataset
        # total_files_max = num_base_files + num_poisoned_files

        # # Calculate how many files are needed from each directory to maintain the split percentage
        # num_poisoned_needed = round(total_files_max * split_percentage)
        # num_base_needed = total_files_max - num_poisoned_needed
        # num_base_needed = min(num_base_needed, num_base_files)

        # # Adjust if not enough poisoned files are available
        # if num_poisoned_needed > num_poisoned_files:
        #     num_poisoned_needed = num_poisoned_files
        #     num_base_needed = total_files_max - num_poisoned_needed

        # # Select files to copy
        # files_to_copy = random.sample(just_poisoned_files, num_poisoned_needed)
        # files_to_copy.extend(random.sample(base_data_files, num_base_needed))

        # # Print final statistics
        # actual_poisoned_count = len([f for f in files_to_copy if f in just_poisoned_files])
        # actual_split = (actual_poisoned_count / len(files_to_copy)) * 100 if files_to_copy else 0

        final_dataset_size = num_base_files

        num_poisoned_needed = min(round(final_dataset_size * split_percentage), num_poisoned_files)
        num_base_needed = final_dataset_size - num_poisoned_needed

        files_to_copy = random.sample(just_poisoned_files, num_poisoned_needed)
        files_to_copy.extend(random.sample(base_data_files, num_base_needed))

        actual_poisoned_count = len([f for f in files_to_copy if f in just_poisoned_files])
        actual_split = (actual_poisoned_count / len(files_to_copy)) * 100 if files_to_copy else 0

        print(f"Total files in final dataset: {len(files_to_copy)}")
        print(f"Number of files copied from just poisoned data: {actual_poisoned_count}")
        print(f"Poison split: {actual_split:.2f}% for label '{label}'")

        
        for file in files_to_copy:
            shutil.copyfile(file, os.path.join(full_dst_path, os.path.split(file)[1]))
```
