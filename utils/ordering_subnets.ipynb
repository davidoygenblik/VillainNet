```python
import pickle
```

```python
def sort_with_index(data):
    """
    Sorts a list while keeping track of the original indices.

    Args:
        data: The list to be sorted.

    Returns:
        A tuple containing the sorted list and a list of original indices.
    """
    indexed_data = list(enumerate(data))
    indexed_data.sort(key=lambda x: x[1])
    sorted_values = [value for index, value in indexed_data]
    original_indices = [index for index, value in indexed_data]
    return sorted_values, original_indices
```

```python
with open('graph_data/gtsrb_dataset/gtsrb_mobilenet_poison_finetune_medium_subnet_fd.pickle', 'rb') as f:
    asrs = pickle.load(f)
    latencies = pickle.load(f)
    params = pickle.load(f)
    flops = pickle.load(f)
    subnets = pickle.load(f)
    clean_accuracies = pickle.load(f)
    asrs_top5 = pickle.load(f)
    clean_accuracies_top5 = pickle.load(f)

data = {
    "ASRs": asrs,
    "latencies": latencies,
    "params": params,
    "flops": flops,
    "subnets": subnets,
    "clean_accuracies": clean_accuracies,
    "ASRs_top5": asrs_top5,
    "clean_accuracies_top5": clean_accuracies_top5
}

ordered_asrs, og_indexes = sort_with_index(data['ASRs'])
ordered_data = {}
for asr, idx in zip(ordered_asrs, og_indexes):
    ordered_data[asr] = data['subnets'][idx]
```
